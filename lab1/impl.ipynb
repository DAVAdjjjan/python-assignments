```python
import re
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import wordnet
import nltk

nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
nltk.download("averaged_perceptron_tagger")
nltk.download("averaged_perceptron_tagger_eng")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
file_path = './text12.txt'
content: str | None  = None

try:
    with open(file_path) as file:
        content = file.read()

except Exception as e:
    print(f"Виникла помилка: {e}")
```

```python

email_pattern = r"\b([A-Za-z0-9._%+-]+)@([A-Za-z0-9.-]+\.[A-Za-z]{2,})\b"
result = re.sub(email_pattern, r"\1@***", content)

print(result)
```

### a) Підрахунок кількості слів у тексті (без знаків пунктуації та спецсимволів)

```python
words_only = re.findall(r"[a-zA-Z']+", content) if content else []
word_count = len(words_only)

print(word_count)
```

### b) Токенізація слів (з попереднім видаленням email-адрес)

```python
text_no_emails = re.sub(email_pattern, "", content) if content else ""

tokens = word_tokenize(text_no_emails)

print(tokens)
```

### c) Видалення стоп-слів

```python
stop_words = set(stopwords.words("english"))

filtered_tokens = [t for t in tokens if t.lower() not in stop_words]
filtered_text = " ".join(filtered_tokens)

print(filtered_text)
```

### d) Лематизація слів третього речення з визначенням частин мови

```python
def get_wordnet_pos(treebank_tag: str):
    if treebank_tag.startswith("J"):
        return wordnet.ADJ
    elif treebank_tag.startswith("V"):
        return wordnet.VERB
    elif treebank_tag.startswith("N"):
        return wordnet.NOUN
    elif treebank_tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN

sentences = sent_tokenize(text_no_emails)

third_sentence_tokens = word_tokenize(sentences[2])
third_sentence_words = [t for t in third_sentence_tokens if t.isalpha()]

tagged_words = pos_tag(third_sentence_words)

lemmatizer = WordNetLemmatizer()
lemmatized = [(word, tag, lemmatizer.lemmatize(word, get_wordnet_pos(tag))) for word, tag in tagged_words]

print(f"  {'Слово':20s} {'POS':6s} {'Лема':20s}")

for word, tag, lemma in lemmatized:
    print(f"  {word:20s} {tag:6s} {lemma:20s}")
```
